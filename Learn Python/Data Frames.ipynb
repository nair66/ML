```python
###Introduction to Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = 8,4
```

```python
data.head()
```

```python
vis1 = sns.distplot(data.InternetUsers,bins = 30)
```

```python
#Boxplots
vis2 = sns.boxplot(data=data,x='IncomeGroup',y='BirthRate')
```

```python
vis3 = sns.lmplot(data = data,y='BirthRate',x='InternetUsers',fit_reg=False)
```

```python
vis3 = sns.lmplot(data = data,y='BirthRate',x='InternetUsers',fit_reg=False,hue='IncomeGroup',height=7,aspect=1)
```

```python
#Keyword Arguements
```

```python
#marker size:
#The scatter_kws(Keyword arguements) calls the original pyplotlib function with the parameters(dictionary) passed to it.Seaborn does not have a parameter to adjust the size of the marker in its wrapper function
vis3 = sns.lmplot(data = data,y='BirthRate',x='InternetUsers',fit_reg=False,hue='IncomeGroup',height=7,aspect=1,scatter_kws={'s':100})
```
